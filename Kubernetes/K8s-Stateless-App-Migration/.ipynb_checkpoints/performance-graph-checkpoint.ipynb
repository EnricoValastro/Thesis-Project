```python
def mediaTot(number):
   return (number/number)*avgTotalTime
y1 = mediaTot(number)
plt.title("Durata Totale del Processo di Migrazione")
plt.xlabel("Iterazione n.")
plt.ylabel("Tempo medio (s)")
plt.plot(number, time, marker='o', label="Tempo totale")
plt.plot(number,y1, label="AvgTot")
plt.legend()
```

```python
def mediaDown(number):
   return (number/number)*avgDowntime
y2 = mediaDown(number)
plt.title("Downtime del Servizio Durante il Processo di Migrazione")
plt.xlabel("Iterazione n.")
plt.ylabel("Tempo medio (s)")
plt.plot(number, time2, marker='o', color="green", label="Downtime")
plt.plot(number,y2, label="AvgDowntime")
plt.legend()
```
